```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
#Create a simple dataset
student={
    'Hours':[2,4,6,8,10],
    "Marks":[55,60,70,80,88]
}
df=pd.DataFrame(student)
df
```

```python
#Train a linear regression model

model=LinearRegression()
model.fit(df[['Hours']],df['Marks'])
```

```python
print("slop ", model.coef_[0])
print('c ',model.intercept_)
```

```python
#Visualize the best fit line 
plt.scatter(df['Hours'],df['Marks'],color='blue',label="Actual Data")
plt.xlabel('Study Hours')
plt.ylabel("Marks")

plt.legend()
plt.grid(True)
```

```python
#prediction a mark
predict=model.predict(df[['Hours']])
predict
```

```python
#Visualize the best fit line 
plt.scatter(df['Hours'],df['Marks'],color='blue',label="Actual Data")
plt.plot(df['Hours'],predict,color='red',label="Best fit line")
plt.xlabel('Study Hours')
plt.ylabel("Marks")

plt.legend()
plt.grid(True)
```

```python
#predicted a number for this model

pre=model.predict([[7]])[0]
pre
```

```python
#gradient decent 
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.metrics import mean_squared_error
#mean squared error show
mse_linear_regression=mean_squared_error(df['Marks'],predict)
print("Mse for linear Regression",mse_linear_regression)
```

```python
plt.scatter(df['Hours'],df['Marks'],color='blue',label="Actual Data")
plt.plot(df['Hours'],predict,color='red',label="Best fit line")
plt.xlabel('Study Hours')
plt.ylabel("Marks")

plt.legend()
plt.grid(True)
```

```python
#using gradient decent based model linear regression
sgd_Regression=SGDRegressor(
    max_iter=1000,
    learning_rate='invscaling',
    eta0=0.01,
    random_state=42
)
sgd_Regression.fit(df[['Hours']],df['Marks'])

_
```

```python
print("M :",sgd_Regression.coef_[0])
print("C :",sgd_Regression.intercept_[0])
```

```python

prediction_SGD=sgd_Regression.predict(df[['Hours']])
print(prediction_SGD)

pred_sgd=sgd_Regression.predict([[7]])[0]
print(pred_sgd)

#error show in sgd_regression
mse_sgd=mean_squared_error(df['Marks'],prediction_SGD)
mse_sgd
```

```python
#plot of linear regression and sgd_regression

plt.scatter(df['Hours'],df['Marks'],color='blue',label="Actual Data")
plt.plot(df['Hours'],predict,color='green',label="Best fit Linear_Regression")
plt.plot(df['Hours'],prediction_SGD,color='red',label="Best fit SGD_Linear_Regression")

plt.xlabel('Study Hours')
plt.ylabel("Marks")

plt.legend()
plt.grid(True)
```

```python
#show error in linear regression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

mae=mean_absolute_error(df['Marks'],predict)
print("Mean absolute error is :",mae)

mse=mean_squared_error(df['Marks'],predict)
print("mean_squared_error is :", mse)

rmse=np.sqrt(mse)
print("ROOt_mean_square_error is",rmse)
```

```python
#overall see the score of the model is 

r2=r2_score(df['Marks'],predict)

print("Overall score of the model is ",r2)
```
