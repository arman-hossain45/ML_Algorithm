```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
```

```python
# Load Original Training Dataset

df_train = sns.load_dataset("penguins").dropna()
print("Original Dataset Head:\n", df_train.head())

print("\nOriginal Species Count:\n", df_train["species"].value_counts())

# Training Features (NO sex column)
X_train = df_train[["bill_length_mm", "bill_depth_mm","flipper_length_mm", "body_mass_g"]]

# Target
y_train = df_train["species"]
```

HERE THE WE SOLVE 

WE TRAIN ORGINAL DATA SET THEN TEST IT WITH THIS NEW CSV DATA SET WHICH WE CREATE 

```python

# Load Synthetic Testing Dataset
df_test = pd.read_csv("penguins_variation.csv").dropna()

# Testing Features (same columns as training)
X_test = df_test[["bill_length_mm", "bill_depth_mm","flipper_length_mm", "body_mass_g"]]

# Test labels
y_test = df_test["species"]


svm_model = svm.SVC(kernel='linear', C=0.1)
svm_model.fit(X_train, y_train)

print("\nSVM Test Accuracy:")
print(svm_model.score(X_test, y_test))

pred_svm = svm_model.predict(X_test)

print("\nSVM Classification Report:")
print(metrics.classification_report(y_test, pred_svm))

print("\nSVM Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_svm))
```

```python
# 2. Decision Tree MODEL

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

print("\nDecision Tree Test Accuracy:")
print(dt_model.score(X_test, y_test))

pred_dt = dt_model.predict(X_test)

print("\nDecision Tree Classification Report:")
print(metrics.classification_report(y_test, pred_dt))

print("\nDecision Tree Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_dt))
```

```python
# 3. Random Forest MODEL

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

print("\nRandom Forest Test Accuracy:")
print(rf_model.score(X_test, y_test))

pred_rf = rf_model.predict(X_test)

print("\nRandom Forest Classification Report:")
print(metrics.classification_report(y_test, pred_rf))

print("\nRandom Forest Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_rf))
```

```python

# 4. KNN MODEL

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

print("\nKNN Test Accuracy:")
print(knn_model.score(X_test, y_test))

pred_knn = knn_model.predict(X_test)

print("\nKNN Classification Report:")
print(metrics.classification_report(y_test, pred_knn))

print("\nKNN Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_knn))
```

HERE WE TRAIN THE MODEL WITH CSV DATA SET THEN TEST THE MODEL WITH ORGINAL DATASET

```python


# Load Synthetic Training Dataset

df_train = pd.read_csv("penguins_variation.csv").dropna()
print("Synthetic Dataset Head:\n", df_train.head(10))

print("\nSynthetic Species Count:\n", df_train["species"].value_counts())

# Training Features (NO sex column)
X_train = df_train[["bill_length_mm", "bill_depth_mm",
                    "flipper_length_mm", "body_mass_g"]]

# Target
y_train = df_train["species"]
```

```python

# Load Original Testing Dataset

df_test = sns.load_dataset("penguins").dropna()

# Testing Features (same columns as training)
X_test = df_test[["bill_length_mm", "bill_depth_mm",
                  "flipper_length_mm", "body_mass_g"]]

# Test labels
y_test = df_test["species"]
```

```python

# Visualization (Same Style as Your Example)
sns.FacetGrid(df_train, hue="species", height=7).map(
    plt.scatter, "bill_length_mm", "bill_depth_mm"
).add_legend()
plt.show()
```

```python

# 1. SVM MODEL

svm_model = svm.SVC(kernel='linear', C=0.1)
svm_model.fit(X_train, y_train)

print("\nSVM Test Accuracy:")
print(svm_model.score(X_test, y_test))

pred_svm = svm_model.predict(X_test)

print("\nSVM Classification Report:")
print(metrics.classification_report(y_test, pred_svm))

print("\nSVM Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_svm))
```

```python

# 2. Decision Tree MODEL

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

print("\nDecision Tree Test Accuracy:")
print(dt_model.score(X_test, y_test))

pred_dt = dt_model.predict(X_test)

print("\nDecision Tree Classification Report:")
print(metrics.classification_report(y_test, pred_dt))

print("\nDecision Tree Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_dt))
```

```python


# 3. Random Forest MODEL

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

print("\nRandom Forest Test Accuracy:")
print(rf_model.score(X_test, y_test))

pred_rf = rf_model.predict(X_test)

print("\nRandom Forest Classification Report:")
print(metrics.classification_report(y_test, pred_rf))

print("\nRandom Forest Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_rf))
```

```python


# 4. KNN MODEL

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

print("\nKNN Test Accuracy:")
print(knn_model.score(X_test, y_test))

pred_knn = knn_model.predict(X_test)

print("\nKNN Classification Report:")
print(metrics.classification_report(y_test, pred_knn))

print("\nKNN Confusion Matrix:")
print(metrics.confusion_matrix(y_test, pred_knn))
```
