```python
import numpy as np
import pandas as pd 

data={
    "room_size":[500,1000,1500,2500],
    "Bsth_size":[40,60,100,120]
}

df=pd.DataFrame(data)
df
```

```python

covarience_mat=np.cov(df,rowvar=False)
covarience_mat
```

```python
print(type(covarience_mat))
```

```python
eigen_values,eigen_vector=np.linalg.eig(covarience_mat)

print(eigen_values)

print(eigen_vector)
```
