```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs,make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN,AgglomerativeClustering#for hierarchy 
from sklearn.neighbors import NearestNeighbors

from scipy.cluster.hierarchy import dendrogram,linkage
#
```

```python
#ploy the moon data sets 

x,y=make_moons(n_samples=300,noise=0.06,random_state=42)

plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1])
plt.title("originals Dataset(Two Moons)")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

# Kmeans diye ata ke clustering kora possible na tai amra dbscan use korbo

Its mandatory to scalling the data set

```python
scaler=StandardScaler()
x_scale=scaler.fit_transform(x)

plt.figure(figsize=(6,6))
plt.scatter(x_scale[:,0],x_scale[:,1])
plt.title("originals Dataset(Two Moons)")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

# DBSCAN IMPLEMENTATION

```python
#dbscan implementation

dbscan=DBSCAN(eps=0.3,min_samples=5)
db_labels=dbscan.fit_predict(x_scale)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x_scale[:,0],x_scale[:,1],c=db_labels,cmap='viridis')
plt.title("DBscan clustering Result")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()

print("Unique cluster labels ",np.unique(db_labels))
```

```python
#choosing eps using k-distance Plot

neighbors=NearestNeighbors(n_neighbors=5)#closet 5 neighbour find out korbe 
neighbors_fit=neighbors.fit(x_scale)
distance,indices=neighbors_fit.kneighbors(x_scale)
distance=np.sort(distance[:,4])
```

```python
plt.figure(figsize=(6,6))
plt.plot(distance)
plt.title("k distance graph(k=5)")
plt.xlabel("points sorted by distance")
plt.ylabel("5th Nearest Neighbor Distance")
plt.grid()
plt.show()
```

```python
#Hierarchical clustering algorithom implementation

hc=AgglomerativeClustering(n_clusters=6,linkage='ward')
hc_labels=hc.fit_predict(x_scale)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x_scale[:,0],x_scale[:,1],c=hc_labels,cmap='plasma')
plt.title("Hierarchical  clustering Result")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

Dendrogram constraction

```python
linked=linkage(x_scale,method='ward')

plt.figure(figsize=(20,10))
dendrogram(linked)
plt.title("Denrogram (Ward Linkage)")
plt.xlabel("data points")
plt.ylabel("distance")
plt.show()
```
